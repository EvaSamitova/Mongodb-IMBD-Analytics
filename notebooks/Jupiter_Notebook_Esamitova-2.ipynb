```python
import re
import requests
import pandas as pd
from IPython.core.display import HTML

# URL of the webpage for movies in 2010 on Metacritic
url = 'https://www.metacritic.com/browse/movie/all/all/2010/new/?page=1'

# Headers to mimic a browser visit
headers = {'User-Agent': 'Mozilla/5.0'}

# Send a GET request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Extract the webpage content
    webpage_content = response.text
    
    # Extract movie titles using the appropriate HTML tag and class
    titles = re.findall(r'<div data-title="(.*?)"', webpage_content)
    
    # Extract release dates using the provided regular expression
    release_dates = re.findall(r'<span class="u-text-uppercase">\s*(.*?)\s*<\/span>', webpage_content)
    
    # Extract metascores using the provided regular expression
    metascores = re.findall(r'<div title="Metascore\s(.*?)\s', webpage_content)
    
    # Extract descriptions using the provided regular expression
    descriptions = re.findall(r'<div class="c-finderProductCard_description">\s*<span>(.*?)<\/span>', webpage_content, re.DOTALL)
    
    # Combine the data into a DataFrame with columns for titles, release dates, metascores, and descriptions
    movie_data = list(zip(titles, release_dates, metascores, descriptions))
    df = pd.DataFrame(movie_data, columns=['Title', 'Release Date', 'Metascore', 'Description'])
    
    # Display the DataFrame in HTML format
    display(HTML(df.to_html(index=False)))
    
else:
    print(f"Failed to retrieve data: {response.status_code}")
```
