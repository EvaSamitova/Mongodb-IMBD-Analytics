```python
import requests
import re

# URL of the webpage for movies in 2010 on Metacritic
url = 'https://www.metacritic.com/browse/movie/all/all/2010/new/?page=1'

# Headers to mimic a browser visit
headers = {'User-Agent': 'Mozilla/5.0'}

# Send a GET request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Extract the webpage content
    webpage_content = response.text
    
    # Regular expression to find movie titles within the <div data-title="...">
    movie_titles = re.findall(r'<div data-title="(.*?)"', webpage_content)
    
    # Print the movie titles
    for title in movie_titles:
        print(title)
```

```python
import re
import requests
import pandas as pd
from IPython.core.display import HTML

# URL of the webpage for movies in 2010 on Metacritic
url = 'https://www.metacritic.com/browse/movie/all/all/2010/new/?page=1'

# Headers to mimic a browser visit
headers = {'User-Agent': 'Mozilla/5.0'}

# Send a GET request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Extract the webpage content
    webpage_content = response.text
    
    # Extract movie titles using the appropriate HTML tag and class
    titles = re.findall(r'<div data-title="(.*?)"', webpage_content)
    
    # Extract release dates using the provided regular expression
    release_dates = re.findall(r'<span class="u-text-uppercase">\s*(.*?)\s*<\/span>', webpage_content)
    
    # Combine the data into a DataFrame with columns for titles and release dates
    movie_data = list(zip(titles, release_dates))
    df = pd.DataFrame(movie_data, columns=['Title', 'Release Date'])
    
    # Display the DataFrame in HTML format
    display(HTML(df.to_html(index=False)))
    
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
import re
import requests

# URL of the webpage for movies in 2010 on Metacritic
url = 'https://www.metacritic.com/browse/movie/all/all/2010/new/?page=1'

# Headers to mimic a browser visit
headers = {'User-Agent': 'Mozilla/5.0'}

# Send a GET request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Extract the webpage content
    webpage_content = response.text
    
    # Use your regular expression to search for descriptions inside <div class="c-finderProductCard_description"> and <span> tags
    descriptions = re.findall(r'<div class="c-finderProductCard_description">\s*<span>(.*?)<\/span>', webpage_content, re.DOTALL)
    
    # Output the number of descriptions retrieved
    print(f"Retrieved {len(descriptions)} descriptions\n")
    
    # Print all the descriptions with a separator between each
    for i, desc in enumerate(descriptions, 1):
        print(f"Movie {i}:\n{desc.strip()}\n{'-'*40}\n")  # Add separator line after each description
```
